```python
df=df.drop_duplicates(subset=['product_title','asin'], keep='first')
```

```python
df['product_price']=df['product_price'].str.replace('[^0-9.]','',regex=True).astype(float)
```

```python
df['product_price']=df['product_price'].fillna(df['product_price'].median())
```

```python
df['product_original_price']
```

```python
df['product_title']=df['product_title'].str.lower().str.replace('\s+',' ',regex=True).str.strip()
```

```python
mediana=df['product_price'].median()
mediana
```

```python
df['product_original_price']=df['product_original_price'].str.replace('[^0-9.]','',regex=True).astype(float)
```

```python
df=df.drop(columns=['currency'])
```

```python
mm=df['product_star_rating'].mean()
mm
```

```python
df['product_star_rating']=df['product_star_rating'].fillna(df['product_star_rating'].mean())
```

```python
df['product_minimum_offer_price']=df['product_minimum_offer_price'].str.replace('[^0-9.]','',regex=True).astype(float)
```

```python
df['product_minimum_offer_price']=df['product_minimum_offer_price'].fillna(df['product_minimum_offer_price'].median())
```

```python
df['sales_volume']
```

```python
df=df.drop(columns=['delivery'])
```

```python
extracted = df["sales_volume"].str.extract(r"(\d+)\s*([kK])?")
```

```python
df["sales_volume_clean"] = extracted[0].astype(float)
```

```python
df.loc[extracted[1].notna(), "sales_volume_clean"] *= 1000
```

```python
df['has_discounts']=df['product_original_price'].notna()
```
